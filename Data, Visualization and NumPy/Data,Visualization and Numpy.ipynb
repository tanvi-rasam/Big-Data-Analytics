```python
import pandas as pd
from pandas import DataFrame
```

```python
# Q1 -  Create a pandas data frame using the data in the ‘auto-mpg.csv’ file
df_assg1 = pd.read_csv('auto-mpg.csv')
df_assg1
```

```python
# Q2 - Display the first 7 rows in the data frame
#default for head() is first 5 values
print(df_assg1.head(7))
```

```python
#Q4 -  Display the unique values of the ‘cylinders’ field
# From the .info() method it can be inferred that cylinders field is non-null thus we can directly apply the .unique() method
df_assg1['cylinders'].unique()
```

```python
#to print the unique values
print("Unique Values in Cylinders Field")
for x in df_assg1['cylinders'].unique():
    print(x)
```

```python
#Q5 - Show the mean, std, min, and max of 'mpg' attribute
#Mean
df_assg1['mpg'].mean()
```

```python
#Standard Deviation
df_assg1['mpg'].std()
```

```python
#Min
df_assg1['mpg'].min()
```

```python
# Max
df_assg1['mpg'].max()
```

```python
#Q6 - Plot the histogram of 'mpg' field
# Getting the co-ordinates
import numpy as np
np.histogram(df_assg1["mpg"])
```

```python
#Plotting the histogram
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel('mpg ranges', fontsize=10)
plt.ylabel('mpg values', fontsize=10)
ax = plt.hist(df_assg1["mpg"])
```

```python
#Q7 -  Create a new data frame that groups the entries by ‘cylinders’ and aggregates them by taking the mean for all the fields. Display the new data frame
# Create a groupby object
name = df_assg1.groupby('cylinders')

# Apply the mean function to the groupby object
df1_assg1 = name.mean()
print(df1_assg1)
```

```python
#Q8 - Display the index of the new data frame
print(df1_assg1.index)
```

```python
# displaying index
print("Index of df1_assg1- ",df1_assg1.index.name)
for x in df1_assg1.index:
    print(x)
```

```python
# Q9 - Create a new field ‘disp_per_cyl’ that gives the displacement per cylinder

df1_assg1['disp_per_cyl']=df1_assg1['displacement']/df1_assg1.index
df1_assg1
```

```python
# 2nd method
df2 = df1_assg1.assign( disp_per_cyl= df1_assg1['displacement']/df1_assg1.index ) 
df2
```

```python
#Q10 -  Create a bar plot for ‘mpg’ vs ‘cylinders’
df1_assg1['mpg'].plot.bar()
plt.xlabel('Cylinders', fontsize=15)
plt.xticks(rotation='horizontal')
plt.ylabel('mpg', fontsize=15)
plt.title('mpg vs Cylinders')
plt.show()
```

```python
#Q11 - Create a line plot for ‘disp per cyl’ vs ‘cylinders

df1_assg1['disp_per_cyl'].plot.line()
plt.xlabel('Cylinders', fontsize=15)
plt.xticks(rotation='horizontal')
plt.ylabel('disp per cyl', fontsize=15)
plt.title('disp per cyl vs Cylinders')
plt.show()
```
